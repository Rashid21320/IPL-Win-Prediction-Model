```python
total_score_df =delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df
```

```python
# Extract the First inning score
total_score_df=total_score_df[total_score_df['inning']==1]
```

```python
total_score_df
```

```python
# Merging the two DataFrame 
match_df=matches.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df
```

```python
match_df.shape
```

```python
match_df['team1'].unique()
```

```python
teams=[
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Kolkata Knight Riders',
    'Royal Challengers Bangalore',
    'Kings XI Punjab',
    'Delhi Capitals',
    'Chennai Super Kings',
    'Rajasthan Royals'
    
]
    
```

```python
# Replace the name of the team
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capital')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capital')
```

```python
match_df['team1']=match_df['team1'].str.replace('Deccan chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan chargers','Sunrisers Hyderabad')
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
match_df.head()
```

```python
match_df.isnull().sum()
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df=match_df[['match_id','city','total_runs','winner']]
```

```python
# Merging two DataFrame 
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
delivery_df
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.shape
```

```python
delivery_df.info()
```

```python
delivery_df['current_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df.head()
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
# Ball left 
delivery_df['ball_left']=126-(delivery_df['over']*6+delivery_df['ball'])
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna('0')
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x =='0' else '1')
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype('int')
wicket=delivery_df.groupby('match_id')['player_dismissed'].cumsum()
delivery_df['wickets'] =10 - wicket
delivery_df.head()
```

```python
#Current run rate = runs/overs
delivery_df['current_rr']=(delivery_df['current_score']*6)/(120-delivery_df['ball_left'])
```

```python
# Requied run rate
delivery_df['Requied_rr']=(delivery_df['runs_left']*6)/delivery_df['ball_left']
```

```python
delivery_df
```

```python
def result(row):
  return 1 if row['batting_team']==row['winner'] else 0
    
```

```python
delivery_df['result'] =delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','runs_left','ball_left','wickets','total_runs_x','current_rr','Requied_rr','result']]
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['ball_left']!=0]
```

```python
final_df.sample()
```

```python
final_df.isnull().sum()
```

```python
# Model training
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
       ('trf',OneHotEncoder(drop='first'),['batting_team','bowling_team','city'] )],
           remainder='passthrough' )           
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

pipe = Pipeline(steps=[
    ('step1', trf),   # your transformer (e.g., StandardScaler, ColumnTransformer, etc.)
    #('step2', RandomForestClassifier(n_estimators=100, random_state=42))
    ('step2', SVC(kernel='rbf', probability=True))
])

pipe.fit(x_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
```

```python
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
cv_scores = cross_val_score(pipe, x, y, cv=5)  # use full dataset (X,Y)
print("\nCross-Validation Scores:", cv_scores)
print("Mean CV Accuracy:", np.mean(cv_scores))
```
